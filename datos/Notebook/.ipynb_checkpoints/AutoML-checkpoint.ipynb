```python
import h2o
import pandas as pd
from h2o.automl import H2OAutoML
```

```python
h2o.init()
```

```python
dataframe = h2o.import_file("vista v3.csv")
dataframe.head()
```

```python
train,test = dataframe.split_frame([0.8],seed=42)
```

```python
target_to_predict = "resultado"
ignore_columns = ["resultado","obesidad","diabetes","enfermedades_respiratorias","hipertension"]
column_to_train = list(set(train.names)-set(ignore_columns))
column_to_train
```

```python
aml = H2OAutoML(max_runtime_secs=7000,seed = 42, project_name = 'classification',balance_classes=True)
%time aml.train(x=column_to_train,y=target_to_predict,training_frame=train)
```

```python
leaderboards = aml.leaderboard
```

```python
leaderboards
```

```python
model = h2o.get_model('GBM_grid_1_AutoML_1_20211205_173237_model_87')
perfomance =model.model_performance(test)
perfomance
```
